```python
import pandas as pd
```

```python
data  = pd.read_csv("./data/movie_metadata.csv")
data.head(10)
```

```python
# create a df with only relevant columns
df = data[["movie_title","genres", "director_name","duration","language","country","title_year",    "imdb_score","movie_imdb_link"]].copy()
df.head()
```

```python
df.notnull().sum()
```

```python
df["movie_title"][df["movie_title"].duplicated()==True].sort_index()
```

```python
df.drop_duplicates(subset=["movie_title"], inplace=True)
df["movie_title"][df["movie_title"].duplicated()==True].sort_index()
```

```python
df.dtypes
```

```python
df["movie_title"].astype("str")
df["imdb_score"].astype("float")
df["title_year"].astype("int")
```

```python
# fectch all the genres

genres_all_movies = [df.loc[i]["genres"].split('|') for i in df.index]

# find the list of genres of all movies in order
genres = sorted(list(set([genre for sublist in genres_all_movies for genre in sublist])))
genres

```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Split genres into lists
df["genres"] = df["genres"].str.split("|")

# # Multi-hot encoding
mlb = MultiLabelBinarizer()
genre_features = mlb.fit_transform(df["genres"])
genre_df = pd.DataFrame(genre_features, columns=mlb.classes_, index=df.index)


```

```python
import joblib

# Save MultiLabelBinarizer
joblib.dump(mlb, "./pickles/mlb_genre_encoder.joblib")
```

```python
genre_df
```

```python
genre_df.to_csv("./data/genre_data.csv")
df.to_json("./data/movie_data.json")
```
